```python
import pandas as pd
```

```python
# set column types
column_types = {
    "CAMPUS": pd.np.str,
}

# import raw data
raw_2016 = pd.read_csv('../data-raw/csv-any/2016-any-CADV.dat', index_col=None, dtype=column_types)
raw_2015 = pd.read_csv('../data-raw/csv-any/2015-any-CADV.dat', index_col=None, dtype=column_types)
raw_2014 = pd.read_csv('../data-raw/csv-any/2014-any-CADV.dat', index_col=None, dtype=column_types)
raw_2013 = pd.read_csv('../data-raw/csv-any/2013-any-CADV.dat', index_col=None, dtype=column_types)
raw_2012 = pd.read_csv('../data-raw/csv-any/2012-any-COTHR.dat', index_col=None, dtype=column_types)
```

```python
# add year columns
yr_2016 = raw_2016.assign(YEAR = '2016')
yr_2015 = raw_2015.assign(YEAR = '2015')
yr_2014 = raw_2014.assign(YEAR = '2014')
yr_2013 = raw_2013.assign(YEAR = '2013')
yr_2012 = raw_2012.assign(YEAR = '2012')
```
